## Matrix and vector operations

Calculate vector $N \times 1$ vector $\mathbf{y}$ through matrix-vector product as in
\begin{equation}
\mathbf{y} = \mathbf{A} \mathbf{x}
\end{equation}

where
\begin{equation}
\mathbf{A} =
\begin{bmatrix}
a_{11} & \cdots & a_{1M} \\
\vdots &  & \vdots \\
a_{N1} & \cdots & a_{NM}
\end{bmatrix}_{N \times M} =
\begin{bmatrix}
A[1,:] \\
\vdots \\
A[N,:]
\end{bmatrix}_{N \times M} =
\begin{bmatrix}
A[:,1] & \cdots & A[:,M]
\end{bmatrix}_{N \times M}
\end{equation}

and $\mathbf{x}$ is a $M \times 1$ vector.
