## SMA filter

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
# number of data
ndata = 200

# coordinates
x = np.linspace(0., 2*np.pi, ndata)

# noise-free data
data = 0.4*np.sin(x) + 0.5*np.cos(3*x)

# noisy data
data_noisy = data + np.random.normal(loc=0., scale=0.1, size=ndata)
```

```python
fig = plt.figure(figsize=(8,5))

plt.plot(x, data, 'ob', linewidth = 1., label='noise-free')
plt.plot(x, data_noisy, 'ok', linewidth = 1., label='noisy')
plt.ylabel('data', fontsize = 22)
plt.xlabel('x', fontsize = 22)
plt.axis([-0.25,6.5,-1.2,1.2])
plt.legend(loc='best', fontsize=16)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid()

fig.tight_layout()

plt.show()  
```

```python
filtered = np.zeros_like(data)
ma = []
w = input('Size of window ')
for i in range(len(x)):
    if i == len(x)-(w-1):
        print i
        break
#     print i, np.sum(x[i:i+w]), np.sum(x[i:i+w])/w
    ma.append(np.sum(data[i:i+w])/w)
filtered[(w-1)//2:(w-1)//2+len(ma)] = np.array(ma)
# print filtered, filtered.size
```

```python
fig = plt.figure(figsize=(8,5))

plt.plot(x, data, 'ob', linewidth = 2., label='noise-free')
plt.plot(x, data_noisy, 'ok', linewidth = 1., label='noisy')
plt.plot(x, filtered, 'or', linewidth = 2., label='filtered')
plt.ylabel('data', fontsize = 22)
plt.xlabel('x', fontsize = 22)
plt.axis([-0.25,6.5,-1.2,1.2])
plt.legend(loc='best', fontsize=16)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid()

fig.tight_layout()

plt.show()  
```
