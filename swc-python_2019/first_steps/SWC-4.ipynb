```python
import glob
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
print glob.glob('../data/inflammation*.csv')
```

#### Plotting Differences

```python
filenames = sorted(glob.glob('../data/inflammation*.csv'))

data1 = np.loadtxt(filenames[0], delimiter=',')
data2 = np.loadtxt(filenames[1], delimiter=',')

fig = plt.figure(figsize=(10.0, 3.0))

axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

axes1.set_ylabel('average1')
axes1.plot(np.mean(data1, axis=0), drawstyle='steps-mid')

axes2.set_ylabel('average2')
axes2.plot(np.mean(data2, axis=0), drawstyle='steps-mid')

axes3.set_ylabel('diff')
axes3.plot(np.mean(data1, axis=0)-np.mean(data2, axis=0))#, drawstyle='steps-mid')

fig.tight_layout()

plt.show()
```

#### Generate Composite Statistics

```python
filenames = glob.glob('../data/inflammation*.csv')
composite_data = np.zeros((60,40))
for f in filenames:
    # sum each new file's data into composite_data as it's read
    #
    composite_data += np.loadtxt(f, delimiter=',')
# and then divide the composite_data by number of samples
composite_data /= len(filenames)
print composite_data
```

```python
fig = plt.figure(figsize=(10.0, 3.0))

axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

axes1.set_ylabel('average')
axes1.plot(np.mean(composite_data, axis=0), drawstyle='steps-mid')

axes2.set_ylabel('max')
axes2.plot(np.max(composite_data, axis=0), drawstyle='steps-mid')

axes3.set_ylabel('min')
axes3.plot(np.min(composite_data, axis=0), drawstyle='steps-mid')

fig.tight_layout()

plt.show()
```
